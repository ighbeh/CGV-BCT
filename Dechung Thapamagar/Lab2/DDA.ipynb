#### 1. Implement the DDA algorithm as a function that returns x and y coordinate lists.

```python
import matplotlib.pyplot as plt
def dda_line(x1, y1, x2, y2):
    xes,yes =[],[] 
    dx = x2- x1
    dy = y2- y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x1
    y = y1
    for _ in range(steps + 1):
        xes.append(round(x))
        yes.append(round(y))
        x += x_inc
        y += y_inc
    return xes, yes
def plot_dda(x1, y1, x2, y2):
    xes, yes = dda_line(x1, y1, x2, y2)
    plt.figure(figsize=(6, 6))
    plt.plot(xes, yes, marker="o", linestyle="-", color="blue")
    plt.title("DDA Line Drawing Algorithm")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.axis("equal")
    plt.show()
plot_dda(10,15,38,27)
```

 #### 2. Draw lines with different slopes: ùëö < 1,ùëö > 1,horizontal,vertical and negative slope.

```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    xes, yes = [], []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1
    for _ in range(steps + 1):
        xes.append(round(x))
        yes.append(round(y))
        x += x_inc
        y += y_inc
    return xes, yes

# Create a plot
plt.figure(figsize=(8, 8))

# Lines with different slopes
lines = [
    ((2, 2), (10, 5), 'm < 1'),       # Gentle slope
    ((2, 2), (5, 10), 'm > 1'),       # Steep slope
    ((2, 2), (10, 2), 'Horizontal'),  # Horizontal line
    ((5, 2), (5, 10), 'Vertical'),    # Vertical line
    ((10, 10), (2, 2), 'Negative')    # Negative slope
]

for (x1, y1), (x2, y2), label in lines:
    xes, yes = dda_line(x1, y1, x2, y2)
    plt.plot(xes, yes, marker='o', linestyle='-', label=label)

plt.title("Lines with Different Slopes (DDA)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

### 3.Draw a rectangle using DDA.

```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    xes, yes = [], []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1
    for _ in range(steps + 1):
        xes.append(round(x))
        yes.append(round(y))
        x += x_inc
        y += y_inc
    return xes, yes

# Draw rectangle using 2 corners
x1, y1 = 2, 3   # first corner
x2, y2 = 10, 8  # opposite corner

plt.figure(figsize=(6,6))

# Draw 4 sides using DDA
xes, yes = dda_line(x1, y1, x2, y1)  # top
plt.plot(xes, yes, marker='o', color='blue')

xes, yes = dda_line(x2, y1, x2, y2)  # right
plt.plot(xes, yes, marker='o', color='red')

xes, yes = dda_line(x2, y2, x1, y2)  # bottom
plt.plot(xes, yes, marker='o', color='green')

xes, yes = dda_line(x1, y2, x1, y1)  # left
plt.plot(xes, yes, marker='o', color='black')

plt.title("Rectangle using DDA")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axis("equal")
plt.show()
```
