```python
# LAB-5
# Q2: Draw ellipses with different radii and centres

import matplotlib.pyplot as plt

def plot_ellipse_points(xc, yc, x, y, xes, yes):
    for px, py in [
        ( x + xc,  y + yc),
        (-x + xc,  y + yc),
        ( x + xc, -y + yc),
        (-x + xc, -y + yc)
    ]:
        xes.append(px)
        yes.append(py)

def midpoint_ellipse(rx, ry, xc, yc):
    rx2, ry2 = rx**2, ry**2
    x, y = 0, ry
    xes, yes = [], []

    p1 = ry2 - rx2 * ry + 0.25 * rx2
    plot_ellipse_points(xc, yc, x, y, xes, yes)

    while 2 * ry2 * x <= 2 * rx2 * y:
        x += 1
        if p1 < 0:
            p1 += 2 * ry2 * x + ry2
        else:
            y -= 1
            p1 += 2 * ry2 * x - 2 * rx2 * y + ry2
        plot_ellipse_points(xc, yc, x, y, xes, yes)

    p2 = ry2 * (x + 0.5)**2 + rx2 * (y - 1)**2 - rx2 * ry2
    while y >= 0:
        y -= 1
        if p2 > 0:
            p2 += rx2 - 2 * rx2 * y
        else:
            x += 1
            p2 += 2 * ry2 * x - 2 * rx2 * y + rx2
        plot_ellipse_points(xc, yc, x, y, xes, yes)

    return xes, yes

ellipses = [(350,105,0,0), (200,150,110,50), (400,205,-100,-10)]

plt.figure(figsize=(6,6))
for rx, ry, xc, yc in ellipses:
    x, y = midpoint_ellipse(rx, ry, xc, yc)
    plt.scatter(x, y, s=10, label=f"rx={rx}, ry={ry}")

plt.title("Task 2: Different Ellipses")
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
