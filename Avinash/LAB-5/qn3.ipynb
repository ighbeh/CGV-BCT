```python
# LAB-5
# Q3: Compare the point spacing in Region 1 and Region 2

import matplotlib.pyplot as plt

def plot_ellipse_points(xc, yc, x, y, xes, yes):
    for px, py in [
        ( x + xc,  y + yc),
        (-x + xc,  y + yc),
        ( x + xc, -y + yc),
        (-x + xc, -y + yc)
    ]:
        xes.append(px)
        yes.append(py)

def midpoint_ellipse_regions(rx, ry):
    rx2, ry2 = rx**2, ry**2
    x, y = 0, ry

    x1, y1 = [], []
    x2, y2 = [], []

    p1 = ry2 - rx2 * ry + 0.25 * rx2
    plot_ellipse_points(0, 0, x, y, x1, y1)

    while 2 * ry2 * x <= 2 * rx2 * y:
        x += 1
        if p1 < 0:
            p1 += 2 * ry2 * x + ry2
        else:
            y -= 1
            p1 += 2 * ry2 * x - 2 * rx2 * y + ry2
        plot_ellipse_points(0, 0, x, y, x1, y1)

    p2 = ry2 * (x + 0.5)**2 + rx2 * (y - 1)**2 - rx2 * ry2
    while y >= 0:
        y -= 1
        if p2 > 0:
            p2 += rx2 - 2 * rx2 * y
        else:
            x += 1
            p2 += 2 * ry2 * x - 2 * rx2 * y + rx2
        plot_ellipse_points(0, 0, x, y, x2, y2)

    return x1, y1, x2, y2


r1x, r1y, r2x, r2y = midpoint_ellipse_regions(300, 150)

plt.figure(figsize=(6,6))
plt.scatter(r1x, r1y, s=10, label="Region 1")
plt.scatter(r2x, r2y, s=10, label="Region 2")
plt.title("Task 3: Region 1 vs Region 2 Point Spacing")
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
