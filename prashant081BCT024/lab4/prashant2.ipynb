1. Implement the midpoint circle algorithm as above.

```python
import matplotlib.pyplot as plt

def midpoint_circle(xc, yc, r):
    x = 0
    y = r
    p = 1 - r

    x_points = []
    y_points = []

    while x <= y:
        # 8-way symmetry
        x_points.extend([xc + x, xc - x, xc + x, xc - x,
                         xc + y, xc - y, xc + y, xc - y])
        y_points.extend([yc + y, yc + y, yc - y, yc - y,
                         yc + x, yc + x, yc - x, yc - x])

        x += 1
        if p < 0:
            p = p + 2 * x + 1
        else:
            y -= 1
            p = p + 2 * (x - y) + 1

    plt.scatter(x_points, y_points)
    plt.title("Midpoint Circle Algorithm")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.axis("equal")
    plt.grid(True)
    plt.show()

# Example
midpoint_circle(0, 0, 10)
```

2. Draw circles with different radii and centres.

```python
import matplotlib.pyplot as plt

def midpoint_circle(xc, yc, r, x_points, y_points):
    x = 0
    y = r
    p = 1 - r

    while x <= y:
        # 8-way symmetry
        x_points.extend([xc + x, xc - x, xc + x, xc - x,
                         xc + y, xc - y, xc + y, xc - y])
        y_points.extend([yc + y, yc + y, yc - y, yc - y,
                         yc + x, yc + x, yc - x, yc - x])

        x += 1
        if p < 0:
            p = p + 2*x + 1
        else:
            y -= 1
            p = p + 2*(x - y) + 1


# Lists to store all circle points
X, Y = [], []

# Draw circles with different centers and radii
midpoint_circle(0, 0, 10, X, Y)
midpoint_circle(15, 10, 6, X, Y)
midpoint_circle(-12, 8, 4, X, Y)
midpoint_circle(8, -10, 8, X, Y)

# Plot all circles
plt.scatter(X, Y)
plt.title("Midpoint Circle Algorithm: Different Centers and Radii")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.grid(True)
plt.show()
```

3. Use the algorithm to draw concentric circles (target pattern).

```python
import matplotlib.pyplot as plt

def midpoint_circle(xc, yc, r, X, Y):
    x = 0
    y = r
    p = 1 - r

    while x <= y:
        # 8-way symmetry
        X.extend([xc + x, xc - x, xc + x, xc - x,
                  xc + y, xc - y, xc + y, xc - y])
        Y.extend([yc + y, yc + y, yc - y, yc - y,
                  yc + x, yc + x, yc - x, yc - x])

        x += 1
        if p < 0:
            p = p + 2*x + 1
        else:
            y -= 1
            p = p + 2*(x - y) + 1


# Center of concentric circles
xc, yc = 0, 0

# Different radii (target pattern)
radii = [4, 8, 12, 16, 20]

X, Y = [], []

for r in radii:
    midpoint_circle(xc, yc, r, X, Y)

# Plot
plt.scatter(X, Y)
plt.title("Concentric Circles using Midpoint Circle Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.grid(True)
plt.show()
```
